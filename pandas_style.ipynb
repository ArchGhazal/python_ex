```python
import pandas as pd
import numpy as np
```

## Create a dataframe of ten rows, four columns with random values

```python
df = pd.DataFrame(np.random.rand(4,10))
```

```python
df
```

## Convert some values to nan values.

```python
df.iloc[2,3]= np.nan
df.iloc[0, 2] = np.nan
df.iloc[3, 3] = np.nan
```

```python
df
```

## Write a Pandas program which will highlight the nan values

```python
df_style = df.style.highlight_null(null_color='red')
```
